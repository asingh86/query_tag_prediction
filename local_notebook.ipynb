```python
import tensorflow as tf
text_test = ['this is such an amazing movie!']
multiclass_model = bert.build_bert_model()
bert_raw_result = multiclass_model(tf.constant(text_test))
print(bert_raw_result)
```

```python
loss = tf.keras.losses.CategoricalCrossentropy()
metrics = tf.metrics.Accuracy()

epochs = 5
steps_per_epoch = len(x_train)
num_train_steps = steps_per_epoch * epochs
num_warmup_steps = int(0.1*num_train_steps)

init_lr = 3e-5
optimizer = tf.keras.optimizers.Adam(learning_rate=init_lr)

#optimizer = keras.optimization.create_optimizer(init_lr=init_lr,
#                                          num_train_steps=num_train_steps,
#                                          num_warmup_steps=num_warmup_steps,
#                                          optimizer_type='adamw')
```

```python
multiclass_model.compile(optimizer=optimizer,
                         loss=loss,
                         metrics=metrics)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience=3)

history = multiclass_model.fit(x = tf.convert_to_tensor(x_train), y = tf.convert_to_tensor(y_train) , validation_data=(x_val, y_val), epochs=5, callbacks = [early_stopping])
```
